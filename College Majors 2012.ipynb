Exploring data sets using graphs. Only a brief analysis of some of the graphs, the major focus of this project is to use graphs to explore a dataset. Some code will be commented out due to have excess amounts of graphs that lack a textual analysis.

```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline

recent_grads = pd.read_csv('recent_grads.csv', encoding='utf-8')
```

```python
raw_data_count = len(recent_grads)
recent_grads = recent_grads.dropna()
cleaned_data_count = len(recent_grads)

print(f'RAW: {raw_data_count}')
print(f'CLN: {cleaned_data_count}')
```

```python
ss_median = recent_grads.plot(x='Sample_size', y='Median', kind='scatter')
ss_median.set_xlabel('Sample Size')
```

```python
ss_unemployment = recent_grads.plot(x='Sample_size', y='Unemployment_rate', kind='scatter')
ss_unemployment.set_xlabel('Sample Size')
ss_unemployment.set_ylabel('Unemployment Rate')
```

From the two graphs above we can see that that the more popular the major(High sample size) the more average the job(Middle of the pack in both graphs). Both graphs start off chaotically due to low sample sizes of most majors and as the popularity of the major increases it converges towards the middle of the chaotic points. <br><br>

```python
uem_hist = recent_grads['Unemployment_rate'].hist(bins=20, range=(0, 0.2))
uem_hist.set_xlabel('Unemployment Rate')
```

```python
sw_hist = recent_grads['ShareWomen'].hist(bins=20, range=(0, 1)) # bins = 2
sw_hist.set_xlabel('Majors - % Women Graduates')
```

From the chart above it seems like over half the programs are predominantly female, at a glance it looks like a 60-40 split in favor of females. If we split it straight down the middle(bins is set to 2, so it splits at 50%) we see that the right side (50<) is slightly below 100 while the left side (50>) is slightly below 80.<br><br> Female Majority ~95<br>Male Majority ~75

```python
wm_hist = recent_grads['Women'].plot.hist(bins=20, range=(0, 200000), ylim=(0,120))
wm_hist.set_xlabel('Women in program')
```

```python
mn_hist = recent_grads['Men'].plot.hist(bins=20, range=(0, 200000), ylim=(0,120))
mn_hist.set_xlabel('Men in program')
```

```python
md_hist = recent_grads['Median'].hist(bins=20, range=(0,120000))
md_hist.set_xlabel('Salary')
```

The most common median salary for a college graduate is between 30-40k with ~40% of college grads falling into this category. For grads that don't fall into this category, they tend to fall into the > avg category with only less than 15% falling below the average.

```python
# taking a look a closer look at the median salary
md_hist2 = recent_grads['Median'].hist(bins=10, range=(30000,40000))
md_hist2.set_xlabel('Salary')
```

Of the grads that fall into common median ~1/3 of grads have a salary of 35-36k.
